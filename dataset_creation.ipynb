```python
import regex as re
```

## 1. Removing dot from Mr and Mrs

<font size=4, color='green'> This step is required to make sure the paragraph is sliced into sentences correctly

```python
xx_v2 = re.sub(r'Mr\.', 'Mr', str(xx), flags=re.IGNORECASE)
xx_v2 = re.sub(r'Mrs\.', 'Mrs', xx_v2, flags=re.IGNORECASE)
```

## 2. Removing some special characters.
<font size=4, color='green'> When the book was converted from pdf to txt, some of the special symbols changed into a text code like--  


\\xe2\\x80\\x94    -->  " - "  
\\xe2\\x80\\x99    -->  " ' "  
\\xe2\\x80\\xa6    -->  " ... "  

So, these were detected and removed using regular expressions


```python
# apart from the special characters mentioned above, \\n and \\t were also filtered
xx_v3 = re.sub(r"\\xe2\\x80\\x94|\\xe2\\x80\\xa6|\\xe2\\x80\\x98|\\n|\\t", " ", xx_v2, flags=re.IGNORECASE)
```

## 3. Repairing the data

<font size=4, color='green'> A novel contains a lot of quoted text said by different characters. They were also converted to some text-code while converting the pdf to text. 

<font size=4> These were then detected and replaced to their original form i.e  
  \\\xe2\\x80\\x9c --> "  
  \\\xe2\\x80\\x9d --> "

```python
xx_v4 = re.sub(r"\\xe2\\x80\\x9c", '"', xx_v3, flags=re.IGNORECASE)
xx_v4 = re.sub(r"\\xe2\\x80\\x9d", '"', xx_v4, flags=re.IGNORECASE)
```

# 4. Spliting the data on full stop

```python
dat_v4 = xx_v4.split(".")
```

## 5. Retaining the apostrophe

<font size=4, color='green'> while converting the data, the apostrophe got converted into some text-code, so retaining the apostrophe using regular expressions.

```python
for i in dat_v4:
    i = re.sub(r"\\xe2\\x80\\x99", r"'", i, flags=re.IGNORECASE)
    i = re.sub(r" +", " ", i)
    print(i)
```

## Final- Extracting the sentences

<font size=4, color='green'> In a novel, these are lot of instances when a characters says something which is written in quoted text follwed by this sentence " said the character".  
For example--> "Cmon Man" said Harry.    
These type of sentence needs to be removed because they are half sentences and also if there were full and complete sentences inside quote, they were already splitted in the above splitting step. This step is required to remove the remaining quoted text which are not complete sentence.
  
So, we removed the sentence by identifying the words like - 'said' , 'whispered', 'asked' etc  

****************************************************************************


<font size=4, color='green'> Lastly, after this also, there are questions which are also sentences.

So, we again split it with .split("?") but then retain the question mark to make sure our model understands what is affermative sentence and what is a question sentence

```python
pree = []
for i in dat_v4:
    i = re.sub(r"\\xe2\\x80\\x99", r"'", i, flags=re.IGNORECASE)
    i = re.sub(r" +", " ", i)
    for j in i.split('"'):
        if 'said' in j:
            continue
        if 'whispered' in j:
            continue
        if 'asked' in j:
            continue
        if 'shouted' in j:
            continue
            
        temp = j.split("?")
        if len(temp)==1:
            pree.append(temp[0])
        else:
            for k in range(len(temp)-1):
                pree.append(temp[k]+"?")
            pree.append(temp[-1])
```

### Removing extra spaces

```python
pree_v2  = [line for line in pree if line.strip()]
```

# Text preprocessing for Movie Corpus

```python
import regex as re
```

### Step1


```python
movie_corpus = open("C:/Users/rpris/Google Drive/Project_2/data/pre.txt","r")
v1=[]
for f in movie_corpus:
    #removing ... , - - , --
    if len(re.findall(r"- -|--|\.\.\.", f)) !=0:
        continue
        
    # removing \n
    f = re.sub(r"\n","",f)
    
    #removing \
    f = re.sub(r"\\", "", f)
    
    #removing dot
    f = re.sub(r'Mr\.', 'Mr', f, flags=re.IGNORECASE)
    f = re.sub(r'Mrs\.', 'Mrs', f, flags=re.IGNORECASE)
    
    #removing <u>, </u>
    f = re.sub(r"<\S>|</\S>", "", f)
    v1.append(f)
movie_corpus.close()
```

```python
v1
```

### Splitting on the basis of '.' and '?' but retaining the '?'

```python
pree = []
for i in v1:
    i = re.sub(r" +", " ", i)
    for j in i.split("."):
        if re.findall(r"\Aand",j):
            continue
        temp = j.split("?")
        if len(temp)==1:
            pree.append(temp[0])
        else:
            for k in range(len(temp)-1):
                pree.append(temp[k]+"?")
            pree.append(temp[-1])

pree
```

### Removing extra spaces and filtering sentences for a particular length

```python
pree = [line for line in pree if line.strip()]
pree = [line for line in pree if len(line.split())>2 and len(line.split())<20]
```
